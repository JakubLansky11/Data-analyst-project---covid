```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("countries_covid_pandas.csv", index_col="country")
print(data)
```

```python
last_date = (max(data["date"]))
countries_parameters = data[["population_density_2019", "child_mortality_2019", "median_age_2018", "life_exp_diff", "confirmed_inf", "deaths"]][data["date"] == last_date]
print(countries_parameters.sort_values("deaths"))
```

```python
countries_population_density = data[["population_density_2019", "confirmed_inf", "deaths"]][data["date"] == last_date].sort_values("deaths")
print(countries_population_density)
data["population_density_2019"].head(40).plot(kind="bar")
```

```python
last_date = (max(data["date"]))
last_data = data[["confirmed_inf", "deaths"]][data["date"] == last_date].sort_values("deaths")
print(last_data.head(50))
```
